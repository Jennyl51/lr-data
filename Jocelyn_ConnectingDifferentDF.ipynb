## Goal #1 and #2: Combine these Two Datasets into One & Keep Important Columns

```python
# import all 3 datasets
```

```python
calls_for_service = pd.read_csv("Data/callsforservice.csv")
```

```python
calls_for_service.head(5)
```

```python
arrest_log = pd.read_csv("Data/arrest_log.csv")
```

```python
arrest_log.head(5)
```

```python
# get rid of the unneeded columns within the Arrest Log 
# (ones that overlap with Calls of Service)
```

```python
arrest_log = arrest_log.drop( ["Date_and_Time", "Arrest_Number", "Date_of_Birth", "Eyes", "Subject", "Hair",  ] , axis = "columns")
```

```python
merge = calls_for_service.merge(arrest_log, left_on = "Incident_Number" , right_on = "Case_Number" )
```

```python
merge.head()
```

```python
merge.to_csv('merged.csv', index=True)
```
